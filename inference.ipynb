```python
from muller_pytorch import MullerResizer, inference
import cv2
import numpy as np
import matplotlib.pyplot as plt

_CONFIGS = {
  "target_size": (512, 512),
 
  "antialias": True,
  "kernel_size": 5,
  "stddev": 1.0,
  "num_layers": 2,
  "avg_pool": False,
 
  "name": 'muller_resizer',
  "init_weights": [1.892, -0.014, -11.295, 0.003],  # pre-trained weights for ResNet-50
}

 

 

model = MullerResizer(**_CONFIGS)
image_file = '/data1/ai_model/panda.jpg'
image = cv2.imread(image_file)

 
pred_image = inference(image, model)
# init model
pred_image = np.transpose(pred_image, (1,2,0)) 
#@title show images
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))
input_image = np.asarray(cv2.cvtColor(cv2.imread(image_file), cv2.COLOR_BGR2RGB)) / 255.0
axes[0].imshow(input_image, aspect="auto")
axes[1].imshow(pred_image, aspect="auto")
```
